# Unión y filtrado de datos

- Métodos de unión:

    - concat(): nos permite unir dataframes por filas y columnas especificando el eje. 
    axis=0 --> pone un dataframe encima del otro. Si tienen el mismo nombre nos pone los datos de esas columnas en la misma columna
    axis=1 --> un dataframe al lado de otro
        - ignore_index=True --> crea un nuevo índice consecutivo. Ojo con los índices

    - join(): al menos una de las columnas de unión tiene que ser un índice.
        - por defecto el how es left
        - si usamos on, la columna de unión tiene que ser del dataframe sobre el que estamos haciendo el join, en el otro utiliza siempre el índice

    - merge(): une dataframes por columnas.

        - por defecto hace inner

- Limpieza:

    - drop(): eliminamos filas o clolumnas en función del eje
        - columns= le podemos pasar una lista de columnas sin especificar el eje
        - inplace=True --> sobreescribe el dataframe. OJO: no igualamos a una variable 

    - rename(): nos permite renombrar columnas.

        - columns= diccionario con el nombre original y el nuevo como value
        - inplace=True --> necesario para sobreescribir el dataframe
    
    - .str(): nos permite usar métodos de strings (.lower(), .split(), ...)

        - .lower() --> tranformar en minúsculas
        - .split() --> dividimos el string por un patron dado. Nos devuelve una lista con los elementos separados
            - expand=True --> devuelve los elemntos en un dataframe con tantas columnas cómo número de elementos tenga la lista creada con el split
                - .get() --> podemos seleccionar las columnas que queramos después del expand
        - .replace() --> reemplazamos valores por otros
        
- Filtrado:

    - Sintaxis:
        ```python
        df[df[columna] operador valor]
        ```

        ```python
        df[df["col"] == 17]
        ```

        ```python
        df[máscara]
        ```

    - .isin(): podemos filtrar por una lista de valores 

        ```python
        df[col].isin([val1, val2])
        ```   

    - .between(): filtrar por intervalos

        ```python
        df[col].between(val_inicial, valor_final)
        ```

        - inclusive= podemos decidir qué partes del intervalo se incluyen: ['both', 'neither', 'left', 'right']

    - .conteins(): podemos filtrar los datos por un patrón

````python
df1 = pd.DataFrame({"A" :[1,2,3], "B":[4,5,6]})
df2 = pd.DataFrame({"B" :[7,8,9], "C":[10,11,12]}, index=[0,1,3])

display(df1)
display(df2)
````

````python
pd.concat([df1,df2], axis = 1)
````

````python
df1.join(df2, lsuffix="1", rsuffix="2")
````

````python
df1.join(df2, on="A", lsuffix="1", rsuffix="2")
````
