# EDA

- Métodos de exploración:

    - .info() --> muestra las columnas, cantidad de filas y columnas, cantidad datos no nulos y el tipo de datos. Al final nos muestra en cantidad de columnas por tipo de dato
    - .describe() --> nos muestra los principales estadísticos (count, media, mediana,...) de las columnas numéricas pr defecto. Si queremos de las columnas catagóricas tenemos que usar include="object" y nos muestra (count, unique, freq,..)
    - .columns --> nos devuelve los nombres de las columnas, nos permite iterar
    - .unique() --> nos devuelve los valores únicos
    - .value_counts() --> conteo de los valores únicos
    - .drop() --> para eliminar filas o columnas. Importante incluir axis(0 para filas y 1 para columnas)
        -   **OJO:**si incluimos el inplace, no podemos igualar a una variable
    - .select_dtypes() --> nos permite filtrar el dataframe por tipo de dato

- Duplicados:

**no siempre es necesario eliminar los valores duplicados**

    - .duplicated() --> nos devuelve una serie de booleanos, para poder ver el total usamos .sum()
    - (keep=): opciones: 
        - "first"(por defecto) nos mantiene el primero y devuelve el resto.
        - "last" mantiene el último y nos devuelve el resto
        - False: nos devuelve todo
    - (subset=): nos permite buscar valores duplicados en una columna
    - .drop_duplicaetes(): funciona igual que .duplicated() en cuanto al keep y subset. Tenemos que añadir inplace=True para que sobreescriba el DF

- Nulos:
    - .isnull()/.isna(): nos devuelve un Dataframe con valores booleanos
    - .isna().sum(): devuelve una serie con los nombres de las columnas y el número de valores nulos. Si añadimos otro .sum() nos da el total de valores nulos del df (ojo: NO el número de filas con valores nulos)
    - Para ver los nulos en porcentaje:
        ```python
        df.isna().sum()/df.shape[0]*100
        ```
        - Para limpiar las columnas que no tienen nulos(solución de Carla):
        ```python
        df.isna().sum()[df.isna().sum() > 0]/df.shape[0]*100
        ```
